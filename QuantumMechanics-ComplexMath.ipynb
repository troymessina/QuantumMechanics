```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sympy import Symbol, exp, I, re, im
from sympy.functions import conjugate
```

```
# In chapter 3, section 1 we encounter a pretty messy complex multiplication
# Let's assume we can write each complex quantity as a real + i*real
# We'll separately deal with the complex exponential, but we know those are
# cos + i*sin from Euler's formula.
a = Symbol('a', positive=True)
b = Symbol('b', positive=True)
c = Symbol('c', positive=True)
d = Symbol('d', positive=True)
f = Symbol('f', positive=True)
g = Symbol('g', positive=True)
k = Symbol('k', positive=True)
m = Symbol('m', positive=True)
# Define E1/hbar = w1 and E2/hbar = w2
w1 = Symbol('w1', positive=True)
w2 = Symbol('w2', positive=True)
t = Symbol('t', positive=True)

alpha1 = a + I*b
alpha2 = c + I*d
c1 = f + I*g
c2 = k + I*m
exp1 = exp(-I*(w2-w1)*t)
exp2 = exp(-I*(w1-w2)*t)
```

```
product1 = alpha1*conjugate(c1)*conjugate(alpha2)*c2*exp1
product2 = conjugate(alpha1)*c1*alpha2*conjugate(c2)*exp2
print(re(product1))
print(re(product2))
```

```
print(im(product1))
print(im(product2))
```

```
print('Real Sum')
print(re(product1)+re(product2))
print('Imaginary Sum')
print(im(product1)+im(product2))
```
