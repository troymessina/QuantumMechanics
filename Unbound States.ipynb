# The Potential Dip
We'll look at the potential dip solutions where a particle with energy above the dip both reflects and transmits. An energy diagram is shown below. We know the solutions are complex exponentials with different wavenumbers in the different regions.

$$   \psi\left(x\right) = 
\left\{
\begin{array}{ll}
      A{\rm e}^{ik_1x}+B{\rm e}^{-ik_1x} & x\lt -a \\
      C{\rm e}^{ik_2x}+D{\rm e}^{-ik_2x} & -a\leq x\leq a \\
      F{\rm e}^{ik_1x} & x\gt a \\
\end{array} 
\right. $$

Analysis of the boundary conditions leads to transmission and reflection probabilities.

\begin{equation}
\begin{split}
T &= \frac{|F|^2}{|A|^2} = \frac{1}{1+\frac{\left(k_1^2-k_2^2\right)^2}{4k_1^2k_2^2}\sin^2\left(2k_2a\right)}
\\
R &= \frac{|B|^2}{|A|^2} = \frac{1}{1+\frac{4k_1^2k_2^2}{\left(k_1^2-k_2^2\right)^2\sin^2\left(2k_2a\right)}}
\\
1 &= T + R
\end{split}
\end{equation}
Since the wavenumbers are related to energy, it is possible to rewrite these probabilities in terms of $E$ and $V_{\circ}$.
\begin{equation}
\begin{split}
T &= \frac{1}{1+\frac{V_{\circ}^2}{4E\left(E+V_{\circ}\right)}\sin^2\left(\frac{2a}{\hbar}\sqrt{2m\left(E+V_{\circ}\right)}\right)}
\\
R &= 1 - T
\end{split}
\end{equation}
Let's plot the transmission and reflection probabilities to explore the results.



```
import matplotlib.pyplot as plt
import numpy as np
#from numpy import *
%matplotlib inline
params = {'legend.fontsize': 'xx-large',
          'figure.figsize': (7.5, 5),
         'axes.labelsize': 'xx-large',
         'axes.titlesize':'xx-large',
         'xtick.labelsize':'xx-large',
         'ytick.labelsize':'xx-large'}
plt.rcParams.update(params)
plt.style.use('ggplot')
#plt.xkcd()
```

```
x_axis = np.arange(-2, 2.001, 0.001)
potential_energy = np.where(x_axis>=-1, -1, 0)
E_total = np.zeros(len(x_axis)) + 0.1
potential_energy = np.where(x_axis>1, 0, potential_energy)
plt.plot(x_axis, potential_energy, label='PE')
plt.plot(x_axis, E_total, label='Total E')
plt.legend(loc=0)
plt.show()
```

```
#We'll assume Vo = -1 eV. This makes E = E/Vo
Vo = -1 # eV
a = 1.0e-9 # meters
hbar = 1.05e-34 # Joule-secs
m = 9.11e-31 # kg of electron
E = np.arange(0.001, 3.001, 0.001) #x axis
transmission = 1/(1+Vo**2/(4*E*(E-Vo))*(np.sin(2*a/hbar*np.sqrt(2*m*(E-Vo)*1.602e-19)))**2)
reflection = 1-transmission

plt.plot(E, transmission, label='T')
plt.plot(E, reflection, label='R')
plt.legend(loc=0)
plt.xlabel('E/Vo')
plt.ylabel('T, R')
plt.show()
```
